```python
import matplotlib.pyplot as plt
from math import factorial as fact

def combination(n,r):
    return (fact(n)/(fact(r)*fact(n-r)))

def binomial(n,success_prob):
    prob = []
    for x in range(n+1):
        p = combination(n,x)*pow(success_prob,x)*pow(1-success_prob,n-x)
        prob.append(p)
        
    """for i in range(n+1):
        print(i,'\t',prob[i])"""
        
    plt.bar(range(n+1),prob)
    plt.show()
    
binomial(10,0.1)
binomial(10,0.9)
binomial(10,0.5)
binomial(100,0.1)
binomial(100,0.9)
binomial(100,0.5)
```

```python
import math

def poisson(l):
    prob = []
    p_x = 1
    x = 0
    while p_x > 0.0000001:
        p_x = math.exp(-l)*pow(l,x)/math.factorial(x);
        prob.append(p_x)
        x = x + 1
    plt.bar(range(x),prob)
    plt.show()
    
poisson(1)
poisson(2)
poisson(10)
```
