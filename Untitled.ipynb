```python
import requests
url = "http://books.toscrape.com/"
page = requests.get(url)
```

```python
print(page.status_code)
```

```python
page.content
```

```python
from bs4 import BeautifulSoup as bs
soup =  bs(page.content,"html.parser")
print(soup)
```

```python
print(soup.prettify())
```

```python
results = soup.find(class="container-fluid page")
print(results.prettify())
```

```python
job_elements = results.find_all(class_ = "card-content")
for job in job_elements:
    print(job.text)
```

```python
job_title = results.find_all(class_ = "title is-5")
for job in job_title:
    print(job.text)
```

```python
jobs = list()
for i in range(0,len(job_title)):
    jobs.append(job_title[i].text)
jobs
```

```python
location =  results.find_all(class_="location")
for loc in location:
    print(loc.text.strip())
```

```python
locations = []
for com_loc in location:
    locations.append(com_loc.text.strip())
locations
```

```python
import pandas as pd
df = pd.DataFrame()
df["Job Title"] = jobs
df["Location"] = locations
df
```
