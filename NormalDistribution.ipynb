```python
import matplotlib.pyplot as plt
import math
```

```python
def normal(mu,sigma):
    prob=[]
    for x in range(-4*sigma+mu,4*sigma+mu):#no. of heads
        p_x=(1/(sigma*math.sqrt(2*math.pi)))*math.exp(-1*pow(((x-mu)/sigma),2)/2)
        prob.append(p_x)
    
    plt.plot(range(-4*sigma+mu,4*sigma+mu),prob)
    plt.show()
```

```python
normal(30,10)
```
