```python
import pandas as pd
df = pd.DataFrame({'date':['2021-01-01','2021-01-02','2021-01-03','2021-01-04','2021-01-01','2021-01-02','2021-01-03','2021-01-04'],'fruit':['apple','apple','apple','apple','mango','mango','mango','mango'],'price':[0.80,None,None,1.20,None,2.10,2.00,1.80]})
df['date'] = pd.to_datetime(df['date'])
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
#number of missing values
```

```python
df[df.isna().any(axis=1)]
#find isna anywhere in the table
```

```python
df[df['price'].isna()]
```

```python
df['price'].fillna(value = 0,inplace=True)
```

```python
df
```

```python
df['price'].fillna(value = df.price.mean(),inplace=True)
```

```python
df
```

```python
df['price'].fillna(value = df.price.median(),inplace=True)
df
```

```python
df['price'].fillna(df.groupby('fruit')['price'].transform('mean'),inplace=True)
df
```

```python
df['price'].fillna(df.groupby('fruit')['price'].transform('median'),inplace=True)
df
```

```python
df.groupby('fruit')['price'].transform('mean')
```

```python
df['price'].fillna(method='ffill',inplace=True)
df
```

```python
df['price'].fillna(method='ffill',limit=1,inplace=True)
df
```

```python
df['price'].fillna(method='bfill',inplace=True)
df
```

```python
df['price'].fillna(method='bfill',limit=1,inplace=True)
df
```

```python
df.groupby('fruit')['price'].ffill(limit=1)
```

```python
df.groupby('fruit')['price'].bfill()
```

```python
df
```

```python
df.groupby('fruit')['price'].ffill().bfill()
```

```python
df.groupby('fruit')['price'].bfill().ffill()
```
