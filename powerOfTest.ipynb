```python
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
from scipy.stats import norm 
```

```python
def power1(n,alpha,sigma,mean):
    mu0 = mean
    z_nd = norm.ppf(alpha)
    se = sigma/sqrt(n)
    critical_xbar = (z_nd * se) + mu0
    power = []

    for i in range(40,mu0):
        z_ad = (critical_xbar - i)/se
        power.append(norm.cdf(z_ad))

    plt.plot(range(40,mu0),power)
    plt.show()
```

```python
def power2(n_list,alpha,sigma,mean):
    mu0 = mean
    z_nd = norm.ppf(alpha)

    for sample_size in n_list:

        se = sigma/sqrt(sample_size)
        critical_xbar = (z_nd * se) + mu0
        power = []

        for i in range(40,mu0):
            z_ad = (critical_xbar - i)/se
            power.append(norm.cdf(z_ad))
        label_value='n =' + str(sample_size)
        plt.plot(range(40,mu0),power,label=label_value)
        plt.legend(loc="lower left")
    plt.show()
```

```python
power2([10,30,50,100],0.05,3,50)
```

```python
def power3(n,alpha_list,sigma,mean):

    for alpha in alpha_list:
        mu0 = mean
        z_nd = norm.ppf(alpha)
        se = sigma/sqrt(n)
        critical_xbar = (z_nd * se) + mu0
        power = []

        for i in range(40,mu0):
            z_ad = (critical_xbar - i)/se
            power.append(norm.cdf(z_ad))
        label_value='alpha = ' + str(alpha)
        plt.plot(range(40,mu0),power,label=label_value)
        plt.legend(loc="lower left")
    plt.show()
```

```python
power3(50,[0.01,0.05,0.1,0.2],10,50)
```

```python
def power4(n,alpha,sigma_list,mean):

    for sigma in sigma_list:
        mu0 = mean
        z_nd = norm.ppf(alpha)
        se = sigma/sqrt(n)
        critical_xbar = (z_nd * se) + mu0
        power = []

        for i in range(40,mu0):
            z_ad = (critical_xbar - i)/se
            power.append(norm.cdf(z_ad))
        label_value='sigma = ' + str(sigma)
        plt.plot(range(40,mu0),power,label=label_value)
        plt.legend(loc="lower left")
    plt.show()
```

```python
power4(40,0.05,[1,3,5,10,20],50)
```
