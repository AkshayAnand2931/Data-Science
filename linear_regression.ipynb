```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {"depth" : [284.5,407.5,512,551,578.5,697,746.5],"age" : [1255,3390,5560,6670,7160,9820,11030]}
df = pd.DataFrame(data)
df
```

```python
x_bar = np.mean(df['depth'])
y_bar = np.mean(df['age'])
print("The mean of the depth is",x_bar,"and the mean of the age is",y_bar)
```

```python
num = 0
denom = 0
for value in zip(df['depth'],df['age']):
    num = num + (value[0]-x_bar)*(value[1]-y_bar)
    denom = denom + ((value[0]-x_bar)**2)
b1 = num/denom
b0 = y_bar - b1*x_bar
print("The value of beta1 is",b1,"and the value of beta0 is",b0)
```

```python
print("The scatter plot of the given data is")
plt.scatter(df['depth'],df['age'])
plt.show()
```

```python
expected = []
for x in df['depth']:
    expected.append((b1*x)+b0)
print("The expected values of the data is")
print(expected)
```

```python
print("The regression model is")
plt.scatter(df['depth'],df['age'])
plt.plot(df['depth'],expected)
plt.show()
```

```python
error = []
count = 0
for y in df['age']:
    error.append(y-expected[count])
    count+=1
print("The errors in the values are")
print(error)
```

```python
plt.scatter(df['depth'],error)
plt.axhline(y = 0)
plt.show()
```

```python
plt.scatter(df['age'],error)
plt.axhline(y = 0)
plt.show()
```
