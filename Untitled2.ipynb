```python
import numpy as np
sales = np.array([[50,2,100,43,200],[48,5,102,44,205],[42,4,109,52,180],[67,10,120,60,222],[105,79,182,107,352],[176,202,190,120,359],[150,303,140,100,323]])
countries = ['USA','India','Sweden','Germany','Japan']
years = [2016,2017,2018,2019,2020,2021,2022]
fig,ax = plt.subplots()
im = ax.imshow(sales)
ax.set_xticks(np.arange(len(countries)))
ax.set_yticks(np.arange(len(years)))
ax.set_xticklabels(countries)
ax.set_yticklabels(years)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",rotation_mode="anchor")
for i in range(len(years)):
    for j in range(len(countries)):
        text = ax.text(j, i, sales[i, j], ha="center", va="center", color="w")
fig.tight_layout()
plt.show()
```
