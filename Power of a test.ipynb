```python
from math import sqrt
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
n=30
sigma=10
alpha=0.05

# H0 : mu>=50
# H1 : mu<50 // left tailed test ; power:are to the left

mu0=50
```

```python
z_nd=norm.ppf(0.05) #to calculate z score give sig level in this fn 
se=sigma/sqrt(n) # std error
critical_xbar=z_nd*se +mu0 #critical value 
power=[]
for i in range(40,mu0):#range of values for mu1
    z_ad=(critical_xbar-i)/se # z score for alt dist
    power.append(norm.cdf(z_ad)) # to calc are to left of z score -- use norm.cdf ; ar to right =1-norm.cdf
plt.plot(range(40,mu0),power)
plt.show()
```

```python
#changing n values 

n_list=[10,30,100,500]

z_nd=norm.ppf(0.05) 

for sample_size in n_list:
    se=sigma/sqrt(sample_size)
    critical_xbar= z_nd*se + mu0
    
    # calc power for diff alternatives 
    power=[]
    for i in range(40,mu0):
        z_ad=(critical_xbar-i)/se
        power.append(norm.cdf(z_ad))
    label_value="n ="+str(sample_size)
    plt.plot(range(40,mu0),power, label=label_value)
    plt.legend(loc="lower left")
plt.show()
```

```python
#changing sigma values 

sigma_list=[20,30,50,100]

z_nd=norm.ppf(0.05) 

for sample_size in sigma_list:
    se=sigma/sqrt(sample_size)
    critical_xbar= z_nd*se + mu0
    
    # calc power for diff alternatives 
    power=[]
    for i in range(40,mu0):
        z_ad=(critical_xbar-i)/se
        power.append(norm.cdf(z_ad))
    label_value="sigma ="+str(sample_size)
    plt.plot(range(40,mu0),power, label=label_value)
    plt.legend(loc="lower left")
plt.show()
```
